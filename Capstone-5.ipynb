```python
#We cluster the neighbourhood by location. I choose five cluster
from sklearn.cluster import KMeans
import numpy as np


X = df['lat'].values.tolist()
Y = df['lng'].values.tolist()

Z = []
ind = 0
for i in X:
    Z += [[i,Y[ind]]]
    ind += 1

X_final = np.array(Z)
kmeans = KMeans(n_clusters=6, random_state=0).fit(X_final)

position_cluster = kmeans.cluster_centers_

#We display the position of the five clusters
position_cluster
```

```python
from math import radians, cos, sin, asin, sqrt
```

```python
#This functions measure distance between two points (latitude, longitude)
AVG_EARTH_DIAMETER = 12742.0176
def haversine(point1, point2):
    lat1, lng1 = map(radians, point1)
    lat2, lng2 = map(radians, point2)
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = sin(lat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(lng / 2) ** 2
    return AVG_EARTH_DIAMETER * asin(sqrt(d))
```

```python
res_lat = df['lat']
res_lng = df['lng']
res_name = df['name']
#print(res_name)
```

```python
#we use the haversine function to identify the restaurants in a radius of 1 km
ind = 0
nb_neigh_list = []
name_list = []
for i in lat_d:
    #print(i)
    #print(lng_d[ind])
    pt1 = i, lng_d[ind]
    ind2=0
    nb_neigh = 0
    name = []
    for i2 in res_lat:
        #print(i2)
        #print(res_lng[ind2])
        pt2 = i2, res_lng[ind2]
        distance = haversine(pt1, pt2)
        if distance <= 1:
            nb_neigh += 1
            name += [res_name[ind2]]
        ind2+=1
    nb_neigh_list += [nb_neigh]
    name_list += [name]
    ind+=1
```
