```python
latitude = 48.8569212 #coordinate Paris center 
longitude = 2.3411728

venues_map = folium.Map(location=[latitude, longitude], zoom_start=13) # generate map centred around Paris center

# add the Italian restaurants as blue circle markers
for lat, lng, label in zip(df.lat, df.lng, df.id):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)

# display map
venues_map
```

```python
#We cluster the neighbourhood by location. I choose five cluster
from sklearn.cluster import KMeans
import numpy as np


X = df['lat'].values.tolist()
Y = df['lng'].values.tolist()

Z = []
ind = 0
for i in X:
    Z += [[i,Y[ind]]]
    ind += 1

X_final = np.array(Z)
kmeans = KMeans(n_clusters=6, random_state=0).fit(X_final)

position_cluster = kmeans.cluster_centers_

#We display the position of the five clusters
position_cluster
```

```python
#We display the five cluster in the map
for i in position_cluster:
    
    folium.CircleMarker(
    [i[0], i[1]],
    radius=10,
    color='green',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(venues_map) 
```

```python
#We make a list for latitude and longitude of clusters
lat_d = []
lng_d =[]
for i in position_cluster:
    lat_d += [i[0]]
    lng_d += [i[1]]
print(lat_d)
print(lng_d)
```

```python
#We make a dataframe with district latitude and longitude
D_name = ['District 1', 'District 2', 'District 3', 'District 4', 'District 5', 'District 6']
d = {'District': D_name, 'lat_district': lat_d, 'lng_district': lng_d}
df2 = pd.DataFrame(data = d)
df2
```

```python
from math import radians, cos, sin, asin, sqrt
```

```python
#This functions measure distance between two points (latitude, longitude)
AVG_EARTH_DIAMETER = 12742.0176
def haversine(point1, point2):
    lat1, lng1 = map(radians, point1)
    lat2, lng2 = map(radians, point2)
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = sin(lat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(lng / 2) ** 2
    return AVG_EARTH_DIAMETER * asin(sqrt(d))
```

```python
res_lat = df['lat']
res_lng = df['lng']
res_name = df['name']
#print(res_name)
```

```python
#we use the haversine function to identify the restaurants in a radius of 1 km
ind = 0
nb_neigh_list = []
name_list = []
for i in lat_d:
    #print(i)
    #print(lng_d[ind])
    pt1 = i, lng_d[ind]
    ind2=0
    nb_neigh = 0
    name = []
    for i2 in res_lat:
        #print(i2)
        #print(res_lng[ind2])
        pt2 = i2, res_lng[ind2]
        distance = haversine(pt1, pt2)
        if distance <= 1:
            nb_neigh += 1
            name += [res_name[ind2]]
        ind2+=1
    nb_neigh_list += [nb_neigh]
    name_list += [name]
    ind+=1
    
```

```python
print(nb_neigh_list)
print(name_list)
```

```python
nb_array = np.array(nb_neigh_list) #We add the number of neighbours and the corresponding name in a dataframe
df2['Number of neighbours'] = nb_array
name_array = np.array(name_list)
df2['Neighbours name'] = name_array
```

```python
#Here we go to look for the note of restaurants in each district and we do the mean
avg_note = [] 
for i in df3['Neighbours name']:
    #print(i)
    note = []
    for j in i:
        #print(j)
        df4 = df.loc[df['name'] == j]
        z = df4.iloc[0]['Note']
        #print(z)
        if z != 'None':
            z = float(z)
            note += [z]
            #print(note)
    avg = sum(note) / len(note)
    avg_note += [avg]
```
